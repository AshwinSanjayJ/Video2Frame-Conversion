# Capture All the frames from a video and store it

```python
import cv2 as cv

# Open the video file
cap = cv.VideoCapture('E:/Vs_Opencv/practice/output.avi')

if not cap.isOpened():
    print("Error: Unable to open video file.")
    exit()

# Initialize variables
current_frame = 0

# Loop to read frames and display them
while True:
    ret, frame = cap.read()
    if not ret:
        print("End of video.")
        break

    # Display the frame
    cv.imshow("Output", frame)

    # Write the frame to a file
    cv.imwrite("./data/frame" + str(current_frame) + ".jpg", frame)

    current_frame += 1

    # Add a delay (in milliseconds) and check if 'q' is pressed to exit the loop
    if cv.waitKey(25) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
cap.release()
cv.destroyAllWindows()
```

# Capture video from live webcam and Store it frame by frame

```python
import cv2 as cv

# Open the default camera (typically the first camera connected to the system)
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Error: Unable to open webcam.")
    exit()

# Define the codec and create VideoWriter object
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

# Initialize variables
current_frame = 0

# Loop to read frames from webcam, display them, and write to a file
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Unable to read frame from webcam.")
        break

    # Display the frame
    cv.imshow("Output", frame)

    # Write the frame to the output video file
    out.write(frame)

    # Write the frame to a file (optional)
    cv.imwrite("./datas/frame" + str(current_frame) + ".jpg", frame)

    current_frame += 1

    # Add a delay (in milliseconds) and check if 'q' is pressed to exit the loop
    if cv.waitKey(2) & 0xFF == ord('q'):
        break

# Release the video capture and video writer objects
cap.release()
out.release()

# Close all OpenCV windows
cv.destroyAllWindows()
```
